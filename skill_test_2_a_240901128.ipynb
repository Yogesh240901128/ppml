```python
import numpy as np
import matplotlib.pyplot as plt

# Daily steps for 30 days (example data)
steps = np.array([
5000, 6200, 7000, 4500, 8000, 9000, 7500,
6000, 7200, 8100, 6500, 7000, 9200, 10000,
8700, 7600, 6900, 7300, 8200, 9100, 6400,
5800, 7200, 8500, 9000, 9600, 8800, 7900,
8100, 8700
])

# Basic statistics
total_steps = np.sum(steps)
average_steps = np.mean(steps)
max_steps = np.max(steps)
min_steps = np.min(steps)

print("Total steps:", total_steps)
print("Average steps per day:", average_steps)
print("Maximum steps:", max_steps)
print("Minimum steps:", min_steps)

# Identify high activity days (threshold = 8000)
threshold = 8000
high_activity_days = np.where(steps > threshold)

print("Days with high activity:", high_activity_days[0] + 1)

# Trend analysis
difference = np.diff(steps)

print("Daily change in steps:", difference)

# Visualization
days = np.arange(1, 31)

plt.plot(days, steps, marker='o')
plt.title("Daily Step Count Over a Month")
plt.xlabel("Day")
plt.ylabel("Steps")
plt.grid()
plt.show()
```
