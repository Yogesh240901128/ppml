```python
import pandas as pd
import matplotlib.pyplot as plt

# Given data
data = {
'Product': ['Laptop', 'Smartphone', 'Tablet', 'Monitor', 'Keyboard'],
'Price': [1200, 800, 300, 150, 50],
'Quantity': [10, 20, 15, 25, 50]
}

# Create DataFrame
df = pd.DataFrame(data)

print("Inventory Data:\n", df)

# 1. Calculate total value of each product
df['Total_Value'] = df['Price'] * df['Quantity']
print("\nTotal value of each product:\n", df)

# 2. Find average price
print("\nAverage Price:", df['Price'].mean())

# 3. Find product with highest price
print("\nMost expensive product:\n", df.loc[df['Price'].idxmax()])

# 4. Find product with lowest quantity
print("\nLowest quantity product:\n", df.loc[df['Quantity'].idxmin()])

# 5. Total quantity of all products
print("\nTotal quantity:", df['Quantity'].sum())

# Plot price of products
plt.bar(df['Product'], df['Price'])
plt.title("Product Price Comparison")
plt.xlabel("Products")
plt.ylabel("Price")
plt.show()
```
