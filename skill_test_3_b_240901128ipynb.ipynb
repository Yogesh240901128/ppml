```python
import numpy as np

# Stock A price data (5 days × 3 features: Open, High, Low)
stockA = np.array([
[100, 105, 98],
[102, 108, 101],
[104, 110, 103],
[103, 109, 100],
[105, 111, 104]
])

# Stock B price data
stockB = np.array([
[98, 103, 96],
[101, 107, 99],
[103, 109, 101],
[102, 108, 98],
[104, 110, 102]
])

# Element-wise sum
sum_matrix = stockA + stockB

# Element-wise difference
diff_matrix = stockA - stockB

print("Stock A Data:\n", stockA)
print("\nStock B Data:\n", stockB)

print("\nElement-wise Sum:\n", sum_matrix)

print("\nElement-wise Difference:\n", diff_matrix)
```
