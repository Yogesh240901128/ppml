```python
import numpy as np

# Grades of 30 students (rows = students, columns = subjects)
# Subjects: Math, Science, English

grades = np.array([
[85,78,82],[90,88,84],[76,70,75],[88,92,90],[67,65,70],
[91,89,93],[72,68,74],[80,85,78],[95,94,96],[60,62,65],
[78,80,79],[84,86,83],[73,75,72],[89,90,88],[66,64,68],
[92,91,94],[70,72,69],[83,81,85],[87,88,86],[75,77,74],
[82,84,80],[68,70,72],[90,92,89],[77,79,76],[85,87,88],
[93,95,94],[71,73,70],[88,86,87],[79,81,78],[84,82,83]
])

# Subject-wise statistics
avg_subject = np.mean(grades, axis=0)
max_subject = np.max(grades, axis=0)
min_subject = np.min(grades, axis=0)

print("Average marks (Math, Science, English):", avg_subject)
print("Highest marks (Math, Science, English):", max_subject)
print("Lowest marks (Math, Science, English):", min_subject)

# Total and average marks of each student
total_marks = np.sum(grades, axis=1)
avg_marks = np.mean(grades, axis=1)

# Top performing student
top_student = np.argmax(total_marks)

print("Top performing student index:", top_student + 1)
print("Top student total marks:", total_marks[top_student])
print("Top student average:", avg_marks[top_student])
```
