```python
import numpy as np
scores=np.array([78,45,89,67,90,34,56,72,88,49,60,77,82,41,53,69,95,38,74,66])
print("Average:",np.mean(scores))
print("Highest:",np,max(scores))
print("Lowest:",np.min(scores))
print("\nPass/Fail:")
print(np.where(scores>=50,"Pass","Fail"))
grades=np.where(scores>=90,'A',np.where(scores>=75,'B',np.where(scores>=60,'C',np.where(scores>=50,'D','F'))))
print("\nGrades:",grades)
```
