```python
import pandas as pd

# Create a dummy EEG dataset for demonstration purposes
data = {
    'Participant': ['P1', 'P1', 'P1', 'P2', 'P2', 'P2'],
    'Time': [1, 2, 3, 1, 2, 3],
    'Electrode1': [10.5, 11.2, 10.8, 15.1, 14.8, 15.5],
    'Electrode2': [20.1, 20.5, 19.8, 22.3, 22.1, 22.5],
    'Electrode3': [5.3, 5.8, 5.5, 6.0, 6.2, 5.9]
}
dummy_df = pd.DataFrame(data)
dummy_df.to_csv('eeg_data.csv', index=False)

# Load EEG dataset
# Example CSV structure: Participant, Time, Electrode1, Electrode2, Electrode3
data = pd.read_csv("eeg_data.csv")

# Display first few rows
print("EEG Data:")
print(data.head())

# Basic statistical analysis
mean_values = data.mean(numeric_only=True)
variance_values = data.var(numeric_only=True)
peak_values = data.max(numeric_only=True)

print("\nMean values:")
print(mean_values)

print("\nVariance values:")
print(variance_values)

print("\nPeak values:")
print(peak_values)

# Analysis per participant
participant_stats = data.groupby("Participant").mean(numeric_only=True)

print("\nAverage EEG activity per participant:")
print(participant_stats)

# Trend analysis over time
time_trend = data.groupby("Time").mean(numeric_only=True)

print("\nTrend of EEG signals over time:")
print(time_trend)
```
