```python
import numpy as np

# Example impedance values (in complex form)
Z1 = 5 + 0j      # Resistor
Z2 = 0 - 3j      # Capacitor
Z3 = 0 + 4j      # Inductor

impedances = np.array([Z1, Z2, Z3])

# Series impedance
Z_series = np.sum(impedances)

# Parallel impedance
Z_parallel = 1 / np.sum(1 / impedances)

print("Impedances:", impedances)
print("Total impedance (Series):", Z_series)
print("Total impedance (Parallel):", Z_parallel)
```
