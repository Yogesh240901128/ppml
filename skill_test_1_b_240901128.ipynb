```python
import numpy as np

# Define matrix A
A = np.array([[2, 3],
              [1, 2]])

# Define vector b
b = np.array([8, 5])

# Solve Ax = b
x = np.linalg.solve(A, b)

print("Solution vector x:")
print(x)
```
