```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a dummy CSV file for demonstration if it doesn't exist
try:
    df = pd.read_csv("sales_data.csv")
except FileNotFoundError:
    data = {'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May'],
            'Sales': [100, 120, 150, 130, 180]}
    df = pd.DataFrame(data)
    df.to_csv('sales_data.csv', index=False)
    print("Created a dummy 'sales_data.csv' for demonstration.")

# 1. Read the CSV file (this will now succeed or use the dummy df)
df = pd.read_csv("sales_data.csv")

# 2. Display first 5 rows
print("First 5 rows of dataset:")
print(df.head())

# 3. Check for missing values
print("\nMissing values:")
print(df.isnull().sum())

# Handle missing values (fill with 0)
df.fillna(0, inplace=True)

# 4. Visualize monthly sales trend
plt.plot(df['Month'], df['Sales'], marker='o')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid()
plt.show()
```
